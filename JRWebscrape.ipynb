## b. Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.

```python
titles = jobs.findAll('h2')
print(type(titles))

titles[0].text
```

```python
new_titles = []
for title in titles:
    new_titles.append(title.text)

new_titles
```

## c. Finally, extract the companies, locations, and posting dates for each job. For example, the first job has a company of "Payne, Roberts and Davis",
## a location of "Stewartbury, AA", and a posting date of "2021-04-08". Ensure that the text that you extract is clean, meaning no extra spaces or other
## characters at the beginning or end.

```python
companies = jobs.findAll('h3')
print(type(companies))
```

```python
new_com = []
for company in companies:
    new_com.append(company.text)

new_com
```

```python
locations = jobs.findAll('p', attrs={'class':"location"})
print(type(locations))
```

```python
locations[0]
```

```python
new_loc = []
for loc in locations:
    new_loc.append(loc.text.strip())

new_loc
```

```python
dates = jobs.findAll('time')
print(type(dates))
```

```python
new_date = []
for date in dates:
    new_date.append(date.text)

new_date
```
